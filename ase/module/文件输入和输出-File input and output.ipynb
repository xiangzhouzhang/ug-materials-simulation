# 文件输入和输出

[文件输入和输出和可支持的文件格式](https://wiki.fysik.dtu.dk/ase/ase/io/io.html)

```python
from ase import Atoms
from ase.build import fcc111, add_adsorbate
from ase.io import read, write
from ase.visualize import view
```

构建Cu吸附CO的slab

```python
adsorbate = Atoms('CO')
adsorbate[1].z = 1.1
adsorbate.positions
```

```python
a = 3.61
slab = fcc111('Cu', (2, 2, 3), a=a, vacuum=7.0)
slab
```

```python
com = add_adsorbate(slab, adsorbate, 1.8, 'ontop')
```

将结构保存图片

```python
write('data/CO-Cu.png', slab * (3, 3, 1), rotation='10z,-80x')
```

显示图片

```python
from PIL import Image
img = Image.open('data/CO-Cu.png')
img
```

Pov-Ray(Persistence of Vision Raytracer): 使用光线跟踪绘制三维图像的源代码开放软件;

```python
write('data/CO-Cu.pov', slab * (3, 3, 1),  rotation='10z,-80x')
```

一般的XYZ格式不会包含晶胞大小的信息, ASE的XYZ格式却包含了晶胞信息

```python
write('data/CO-Cu.xyz', slab)
```
