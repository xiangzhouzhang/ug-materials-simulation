## 吸附位点查找器

**AdsorbateSiteFinder()**: 返回一个由slab及其吸附位点组成的吸附结构对象;

```python
import pymatgen
from pymatgen import Structure, Lattice, MPRester, Molecule
from pymatgen.analysis.adsorption import *
from pymatgen.core.surface import generate_all_slabs
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
mpr = MPRester() #provide your own API Key
```

**Structure.from_spacegroup()**: 从空间群建立一个simple fcc(简单面心立方结构)

```python
fcc_ni = Structure.from_spacegroup("Fm-3m", Lattice.cubic(3.5), ["Ni"], [[0, 0, 0]])
print(fcc_ni)
```

**generate_all_slabs()**: 生成fcc所有特定的slab

```python
slabs = generate_all_slabs(fcc_ni, max_index=1, min_slab_size=8.0, min_vacuum_size=10.0)
```

```python
%C len(slabs); [slab.miller_index for slab in slabs]  #slab对象具有弥勒属性
```

选择弥勒指数是(111)的slab

```python
ni_111 = [slab for slab in slabs if slab.miller_index==(1,1,1)][0] #一个指数可能对应几个不同的slab
print(ni_111)
```

实例化位点查找器,以slab为实例化参数, 然后用它寻找吸附位点;    

```python
asf_ni_111 = AdsorbateSiteFinder(ni_111)
ads_sites = asf_ni_111.find_adsorption_sites() #字典
```

```python
#ads_sites是一个字典, 字典值是数组对象
%C 10 list(ads_sites); ads_sites['ontop']
```

**pymatgen.analysis.adsorption.plot_slab()**: 可视化吸附结构

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plot_slab(ni_111, ax, scale=0.8, adsorption_sites=True)
plt.show()
```

**generate_adsorption_structures()**: 创建吸附结构;

```python
fig = plt.figure()
ax = fig.add_subplot(111)
adsorbate = Molecule("H", [[0, 0, 0]]) #怎么知道是000
ads_structs = asf_ni_111.generate_adsorption_structures(adsorbate,repeat=[1, 1, 1])
plot_slab(ads_structs[0], ax, scale=0.8, adsorption_sites=False, decay=0.09)
plt.show()
```

## 在不同表面上寻找吸附位点

使用`AdsorbateSiteFinder`在不同表面(结构或弥勒指数不同)上寻找所有可能的吸附位点;  

```python
# %%pass
mats = {"mp-23":(1, 0, 0),   # FCC Ni
        "mp-2":(1, 1, 0),    # FCC Au
        "mp-13":(1, 1, 0),   # BCC Fe
        "mp-33":(0, 0, 1),   # HCP Ru
        "mp-30": (2, 1, 1),  
        "mp-5229":(1, 0, 0)} # Cubic SrTiO3 

fig = plt.figure()
axes = [fig.add_subplot(2, 3, i) for i in range(1, 7)]
for n, (mp_id, m_index) in enumerate(mats.items()):
    struct = mpr.get_structure_by_material_id(mp_id)
    struct = SpacegroupAnalyzer(struct).get_conventional_standard_structure()
    slabs = generate_all_slabs(struct, 1, 5.0, 2.0, center_slab=True)
    #通过列表推导式生成字典, 密勒指数: slab 
    slab_dict = {slab.miller_index: slab for slab in slabs}
    #`from_bulk_and_miller`是类方法, 直接由类调用
    asf = AdsorbateSiteFinder.from_bulk_and_miller(struct, m_index, undercoord_threshold=0.10)
    #n是子图编号, 在每个子图上绘制吸附结构
    plot_slab(asf.slab, axes[n]) 
    
    #吸附结构的吸附位点
    ads_sites = asf.find_adsorption_sites()
    
    #将slab的z轴旋转到弥勒指数方向
    sop = get_rot(asf.slab)
    ads_sites = [sop.operate(ads_site)[:2].tolist() for ads_site in ads_sites["all"]]
    
    #绘制吸附点
    axes[n].plot(*zip(*ads_sites), color='k', marker='x', markersize=10, mew=1, linestyle='', zorder=10000)
    
    #标题: {结构.组成.化学式}({密勒指数})
    mi_string = "".join([str(i) for i in m_index])
    axes[n].set_title("{}({})".format(struct.composition.reduced_formula, mi_string))
    axes[n].set_xticks([]); axes[n].set_yticks([])    
axes[4].set_xlim(-2, 5); axes[4].set_ylim(-2, 5)
fig.savefig('slabs.png', dpi=200)
```

## 从atomate产生工作流

使用MatMethods, 对确定DFT能量(the determination of), 创建一个完整工作流(full workflow);     
有关工作流的模块: [FireWorks](https://materialsproject.github.io/fireworks/);  非关系数据库模块: [MongoDB](https://docs.mongodb.com/manual/);

```python
from fireworks import LaunchPad
lpad = LaunchPad()
```

```python
lpad.reset('', require_password=False)
```

从atomate导入工作流生成函数

```python
from atomate.vasp.workflows.base.adsorption import get_wf_surface, get_wf_surface_all_slabs
```

吸附配置(Adsorption configurations)字典: {密勒指数: 分子对象};

```python
co = Molecule("CO", [[0, 0, 0], [0, 0, 1.23]])
h = Molecule("H", [[0, 0, 0]])
co
```

Workflows使用slab和一系列分子生成

```python
struct = mpr.get_structure_by_material_id("mp-23") # fcc Ni
struct = SpacegroupAnalyzer(struct).get_conventional_standard_structure()
```

```python
print(struct)
```

```python
slabs = generate_all_slabs(struct, 1, 5.0, 2.0, center_slab=True)
slab_dict = {slab.miller_index:slab for slab in slabs}
ni_slab_111 = slab_dict[(1, 1, 1)] #弥勒指数是(111)的slab
print(ni_slab_111)
```

```python
wf = get_wf_surface([ni_slab_111], molecules=[co, h]) #返回工作流对象(workflow)
lpad.add_wf(wf) #向发射台添加工作流
```

工作流中的任务被视为如下(be inspected as below):    
与slab相关的9个优化任务, 4个不同的吸附构型(点?), distinct adsorption configurations for each of the 2 adsorbates;

## 在二元氧化物中筛选 析氧电催化剂 (Screening of oxygen evolution electrocatalysts on binary oxides)

使用MP-API和吸附工作流, 对过渡金属二元氧化物的电催化吸氧性能进行高通量研究

```python
from pymatgen.core.periodic_table import *
from pymatgen.core.surface import get_symmetrically_distinct_miller_indices
import tqdm
```

```python
lpad.reset('', require_password=False)
```

根据理论过电势(theoretical overpotential), 判断给定催化剂对于吸氧反应的催化活性;  
期间的吸附物质包含: OH\*, O\*,  OOH\*;

```python
OH = Molecule("OH", [[0, 0, 0], [-0.793, 0.384, 0.422]])
O = Molecule("O", [[0, 0, 0]])
OOH = Molecule("OOH", [[0, 0, 0],   #如何判断吸附原子的位置
                       [-1.067, -0.403, 0.796],
                       [-0.696, -0.272, 1.706]])
print(OOH)
```

```python
adsorbates = [OH, O, OOH]
```

使用MP-rest接口获取结构, 循环生成所有的工作流对象: `slab-adsorbate与每种材料的组合;   
该程序将运行大概15分钟; 该示例可以并行处理, 但此处并未涉及;

```python
elements = [Element.from_Z(i) for i in range(1, 103)] # 0-103号元素
```

```python
elements[:6]
```

```python
trans_metals = [el for el in elements if el.is_transition_metal][:3] #为加快速度只取三个元素
```

```python
%omit trans_metals
```

```python
# `tqdm`添加了一个进度条(progress bar), 因此可以看到循环的过程
with open('br_data/ex4.txt','w') as f:
    for metal in tqdm.tqdm_notebook(trans_metals):
        # 获取稳定结构-最小单位晶胞
        data = mpr.get_data("{}-O".format(metal.symbol))
        data = [datum for datum in data if datum["e_above_hull"] < 0.05]
        data = sorted(data, key = lambda x: x["nsites"])
        struct = Structure.from_str(data[0]["cif"], fmt='cif')

        struct = SpacegroupAnalyzer(struct).get_conventional_standard_structure()
        # 获取低指数晶面的特定弥勒指数
        wf = get_wf_surface_all_slabs(struct, adsorbates)
        lpad.add_wf(wf)
        print("Processed: {}".format(struct.formula), file=f)
```
