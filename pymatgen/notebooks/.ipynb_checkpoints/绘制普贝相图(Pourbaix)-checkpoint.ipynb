使用MP-api绘制普贝相图    

结合水性试验和第一性原理预测固液平衡相图:   
[Prediction of solid-aqueous equilibria: Scheme to combine first-principles calculations of solids with experimental aqueous states](https://journals.aps.org/prb/abstract/10.1103/PhysRevB.85.235438);  
亚稳态材料的电化学稳定性:  
[Electrochemical Stability of Metastable Materials](https://pubs.acs.org/doi/10.1021/acs.chemmater.7b03980);

```python
from pymatgen import MPRester
from pymatgen.analysis.pourbaix_diagram import PourbaixDiagram, PourbaixPlotter
%matplotlib inline
```

```python
mpr = MPRester()
```

**MPRester.get_pourbaix_entries(LIST_OF_ENTRIES)**:  
获取普贝相图的相关条目信息, `LIST_OF_ENTRIES`化学系统中的主元素(自动添加H-O);    
该函数会施行所有必要的预处理, 以确保`PourbaixEntries`与`PourbaixDiagram`兼容;

获取与Cu-O-H化学体系对应的普贝条目信息

```python
entries = mpr.get_pourbaix_entries(["Cu"]) #len(entries)=55
```

**PourbaixDiagram(RETRIEVED_ENTRIES)**: 创建普贝相图对象

```python
pbx = PourbaixDiagram(entries)
pbx
```

**PourbaixAnalyzer**: 检测相对于普贝面(facet, pH值和V(电压?)的函数)的稳定相及其稳定性

```python
entry = [e for e in entries if e.entry_id == 'mp-1692'][0]
entry
```

```python
print("CuO's potential energy per atom relative to the most stable decomposition product is",
      "{:0.2f} eV/atom".format(pbx.get_decomposition_energy(entry, pH=7, V=-0.2)))
```

> 这表明CuO在中性pH和温和还原条件下具有很大的分解驱动力

PourbaixPlotter(pbx): 用普贝对象作为参数实例化普贝绘图器

```python
plotter = PourbaixPlotter(pbx)
```

显示普贝相图

```python
plotter.get_pourbaix_plot().show()
```

带稳定性指示的普贝相图

```python
plt = plotter.plot_entry_stability(entry)
plt.show()
```

Pymatgen还支持具有非HO且成分固定的二元/三元pourbaix图;    
寻找满足成分约束的所有组合, 并在普贝空间中将它们视为单独的条目;

```python
entries = mpr.get_pourbaix_entries(["Bi", "V"])
```

```python
%col [entries[i].name for i in range(273)]
```

```python
# Construct the PourbaixDiagram object
pbx = PourbaixDiagram(entries, comp_dict={"Bi": 0.5, "V": 0.5},
                      conc_dict={"Bi": 1e-8, "V": 1e-8}, filter_solids=True)
```

Note that the `filter_solids` keyword argument in the `PourbaixDiagram` instantiation above tells the constructor whether to filter solids by phase stability on the compositional phase diagram.  

Note that Pourbaix Diagrams generated with and without this argument may look significantly different in the OER and HER regions, since highly oxidized materials (e. g. Bi$_2$O$_5$) or highly reduced materials (e. g. most hydrides) may not be stable on the compositional phase diagram. 

The filtering process significantly reduces the time it takes to generate all of the combined entries for a binary or ternary pourbaix diagram though, so it may be prudent to use in those cases.

```python
# Construct the pourbaix analyzer
plotter = PourbaixPlotter(pbx)
plt = plotter.get_pourbaix_plot()
plt.show()
```

Getting the heatmaps for a solid entry in these cases is a bit more involved, because many of the regions of the pourbaix diagram include multiphase entries.  Here's an example for this case.

```python
bivo4_entry = [entry for entry in entries if entry.entry_id=="mp-613172"][0]
plt = plotter.plot_entry_stability(bivo4_entry)
```

If a compound is provided that doesn't meet the constraints of the composition (in this case, equal parts Bi and V), the plotter will attempt to find the most stable entry containing that solid and any spectating ions.  In this case, it estimates the stability of BiO$_2$(s) + VO$_4$-.

```python
bio2_entry = [entry for entry in entries if entry.entry_id=="mp-557993"][0]
plt = plotter.plot_entry_stability(bio2_entry)
```
