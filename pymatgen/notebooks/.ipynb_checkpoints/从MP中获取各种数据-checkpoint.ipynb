# 从MP(Materials Project)获得各种数据

Materials API: https://github.com/materialsproject/mapidoc;

```python
import re
from pymatgen import MPRester, Composition 
%matplotlib inline
```

```python
mpr = MPRester()
```

## 根据MP-ids获得结构

### 寻找与Fe2O3具有相似化学计量(stoichiometry)的所有结构

创建组成对象

```python
comp = Composition("Fe2O3")
```

匿名化学式:A2B3

```python
anon_formula = comp.anonymized_formula 
anon_formula
```

将化学式(formula)转换成`元素字典`形式;

```python
anon_formula = {m.group(1): int(m.group(2)) 
                for m in re.finditer(r"([A-Z]+)(\d+)", anon_formula)} #{'A': 2, 'B': 3}
anon_formula
```

查询

```python
data = mpr.query({"anonymous_formula": anon_formula}, 
                 properties=["task_id", "pretty_formula", "structure", "energy"]) 
len(data)
```

```python
# 查看能量小于-600的结构有哪些
for i, da in enumerate(data):
    if da['energy']<=-600:
        print("{:}({}): {}".format(da['task_id'], i, da['pretty_formula']))
```

## 结构匹配

Pymatgen有自己的结构匹配算法

```python
from pymatgen.analysis.structure_matcher import StructureMatcher
m = StructureMatcher() 
s1 = data[0]["structure"]
s2 = s1.copy()
s2.apply_strain(0.1)
```

```python
%C 1 s1; s2
```

```python
print(m.fit(s1, s2)) 
```

有多少结构与Gd2O3相似

```python
matches = []
for d in data:
    if m.fit_anonymous(d["structure"], s1):
        matches.append(d)
len(matches)
```

```python
# 查看相似结构
%C 1 matches[1]; matches[2] #同构(iso-structural)
```
