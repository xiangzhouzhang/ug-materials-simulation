为**多晶型(polymorphs)CsCl**($Pm\overline{3}m$, $Fm\overline{3}m$)绘制XRD图

```python
from pymatgen import Lattice, Structure
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from IPython.display import Image, display
%matplotlib inline
```

## $\alpha$-CsCl ($Pm\overline{3}m$)

创建alpha-CsCl的晶体结构

```python
a = 4.209 #Angstrom
latt = Lattice.cubic(a)
structure = Structure(latt, ["Cs", "Cl"], [[0, 0, 0], [0.5, 0.5, 0.5]])
print(structure)
```

```python
c = XRDCalculator()
```

```python
c.get_pattern(structure)
```

```python
c.get_plot(structure)
```

## $\beta$-CsCl ($Fm\overline{3}m$)

创建beta-CsCl的晶体结构

```python
a = 6.923 #Angstrom
latt = Lattice.cubic(a)
structure = Structure(latt, ["Cs", "Cs", "Cs", "Cs", "Cl", "Cl", "Cl", "Cl"], 
                          [[0, 0, 0], [0.5, 0.5, 0], [0, 0.5, 0.5], [0.5, 0, 0.5], 
                           [0.5, 0.5, 0.5], [0, 0, 0.5], [0, 0.5, 0], [0.5, 0, 0]])
print(structure)
```

```python
c.get_plot(structure)
```
